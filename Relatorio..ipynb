5.5 Compare a imagem de Y com R, G e B e com Cb e Cr. Tire
conclusões.

![Airport_YcbCr](Ex5_images/Y_airport.png)

Canal Y (Luminância): Este canal contém informações sobre o brilho da imagem. Ele é tipicamente mais detalhado, pois os seres humanos são<br>
mais sensíveis a variações na luminância do que na crominância.<br>

![Airport_YcbCr](Ex5_images/red_airport.png)
![Airport_YcbCr](Ex5_images/green_airport.png)
![Airport_YcbCr](Ex5_images/blue_airport.png)

Canais R, G e B (Cores Primárias): Estes canais representam as cores primárias de luz. O canal vermelho (R) mostrará áreas que são mais<br>
brilhantes onde a imagem tem mais vermelho. O mesmo se aplica aos canais verde (G) e azul (B). Cada um desses canais pode conter uma <br>
quantidade significativa de detalhes, mas eles serão mais representativos das áreas onde sua respectiva cor é predominante na imagem. <br>
Para além disso, nota-se que se pode observar melhor a imagem no canal verde, pois o ser humano tem a visão mais sensível para a cor verde.<br>

![Airport_YcbCr](Ex5_images/Cb_airport.png)
![Airport_YcbCr](Ex5_images/Cr_airport.png)

Canais Cb e Cr (Crominância): Cb representa a diferença entre o azul e a luminância (Y), e Cr representa a diferença<br>
entre o vermelho e a luminância (Y). Esses canais contêm menos detalhes do que o canal de luminância e são geralmente<br>
subamostrados para economizar espaço, pois o olho humano não detecta tanto detalhe nas informações de cor quanto nas<br>
informações de luminância.<br>
